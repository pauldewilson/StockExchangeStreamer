```python
import numpy as np
import string
import time
import json
import shutil
import os
import pyspark
import pyspark.sql.functions as F
from datetime import timedelta
from datetime import datetime as dt
from pyspark import SparkContext, SparkConf, SparkFiles
from pyspark.sql import SparkSession, SQLContext
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, TimestampType, ArrayType
from IPython.display import clear_output
from delta import *
from faker import Faker
import plotly.graph_objects as go
import plotly.express as px
```

```python
builder = pyspark.sql.SparkSession.builder.appName("MyApp") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .config("spark.default.parallelism", "24")

spark = configure_spark_with_delta_pip(builder).getOrCreate()
sc = spark.sparkContext
```

```python
class Environment:

    def __init__(self) -> None:
        self.DIR_ROOT = os.getcwd()+"\sx"
        self.DIR_COMPANIES = os.path.join(self.DIR_ROOT, "companies")
```

```python
ENV = Environment()
```

```python
schema = StructType([
    StructField("ticker", StringType(), True),
    StructField("price", DoubleType(), True),
    StructField("timestamp", TimestampType(), True),
])
```

```python
while True:
    clear_output(wait=True)
    df = spark.read.format("delta").load(os.path.join(ENV.DIR_ROOT, "delta", "trades"))

    df = df.toPandas()

    fig = px.line(df, x="timestamp", y="price", color="ticker")
    fig.show()
    time.sleep(15)
```
